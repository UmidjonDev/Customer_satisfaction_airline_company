```python
#Importing necessary modules and libraries
#Data preprocessing 
import pandas as pd 
import numpy as np 

#Pipeline loader
import dill
```

```python
#Loading model and test dataframe 
df_test = pd.read_csv(filepath_or_buffer = './data/imputed_test_dataset.csv')
pipe = dill.load(open('./pipeline/models/customer_satisfaction.pkl', mode = 'rb'))
print("Model loaded successfully ! ")
df_test.head()
```

```python
model = pipe['model']
result = {
    'id' : df_test['id'],
    'predictions' : model.predict(X = df_test)
}
predictions_df = pd.DataFrame(data = result)
predictions_df.to_excel('./result/my_model_predictions.xlsx', index = False)
print("Submission file is saved successfully in the folder result !")
predictions_df.head(10)
```
